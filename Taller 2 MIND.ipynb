4. Muestre la cantidad de pacientes por cada tipo de tumor (benign o malignant) junto con una representación gráfica.

```python
def graphTumorType():
    df = carga()
    df.groupby('diagnosis')["id"].count().plot(kind = 'bar',color = ['green','red'])
    print(df.groupby('diagnosis')["id"].count())
graphTumorType()
```
